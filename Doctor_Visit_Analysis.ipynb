# Visualize the percentage of people getting PRIVATE health insurance and GOVT health insurance due to low income, old age.


```python
label=['yes','no']
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people PRIVATE health insurance due to low income")
plt.show()

Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people GOVT health insurance due to low income")
plt.show()

Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("Percentage of people GOVT health insurance due to old age")
plt.show()
```

# Plot horizontal bar chart to analyze the reduced days of activity due to illness based on gender


```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
plt.title('HORIZONTAL BAR CHART')
plt.xlabel('REDUCED ACTIVITY')
plt.ylabel('GENDER')
plt.show()
```
